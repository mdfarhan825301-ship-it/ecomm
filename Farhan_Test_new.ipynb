**7.2 Customer Retention Analysis**

```python
retained_statuses = [
    'Shipped',
    'Shipped - Out for Delivery',
    'Shipped - Picked Up',
    'Pending - Waiting for Pick Up',
    'Shipping'
]

# Filter based on valid status
retained_data = df[df['Status'].isin(retained_statuses)]


orders_by_customer_proxy = retained_data.groupby('ship-postal-code')['Order ID'].nunique()

# Count retention metrics
retained_count = (orders_by_customer_proxy > 1).sum()
one_time_count = (orders_by_customer_proxy == 1).sum()
retention_percent = (retained_count / (retained_count + one_time_count)) * 100

print(f"Retained Customers: {retained_count}")
print(f"One-time Customers: {one_time_count}")
print(f"Retention Rate: {retention_percent:.2f}%")
```

**8. Differentiating B2B vs. B2C Customer Behavior**

```python
df['B2B'] = df['B2B'].astype(str).str.lower().map({'true': True, 'false': False})

b2b_b2c_summary = df.groupby('B2B').agg(
    total_orders = ('Order ID', 'nunique'),
    total_revenue = ('Amount','sum'),
    avg_order_value = ('Amount','mean'),
    avg_quantity = ('Qty','mean')
).reset_index()

print("B2B vs B2C Summary:")
print(b2b_b2c_summary)

# Product category preference by B2B vs B2C
category_preference = df.groupby(['B2B','Category']).size().reset_index(name='Order_Count')
print("\nProduct Category Preference by B2B vs B2C:")
print(category_preference.sort_values(['B2B','Order_Count'], ascending=[True,False]).head(10))

b2b_b2c_summary['Revenue_Contribution_%'] = (
    b2b_b2c_summary['total_revenue'] / b2b_b2c_summary['total_revenue'].sum() * 100
)

print("\nRevenue Contribution %:")
print(b2b_b2c_summary[['B2B','Revenue_Contribution_%']])
```

**13. Enhancing Inventory Control to Prevent Stockouts or Overstocking**

```python
# Calculate demand metrics for each product
inventory_control = df.groupby('Category').agg(
    Total_Orders=('Order ID','count'),
    Total_Units_Sold=('Qty','sum'),
    Avg_Qty_Per_Order=('Qty','mean')
).reset_index()

# Classify inventory behavior based on demand
inventory_control['Inventory_Category'] = inventory_control['Total_Units_Sold'].apply(
    lambda x: 'High Demand â€“ Risk of Stockout' if x > inventory_control['Total_Units_Sold'].mean()
              else 'Low Demand â€“ Risk of Overstock'
)

# Sort by units sold to see demand pattern clearly
inventory_control = inventory_control.sort_values(by='Total_Units_Sold', ascending=False)

print("ðŸ“¦ Inventory Control Analysis")
inventory_control
```

# Customer Satisfaction & Returns

**14. Investigating High Return & Cancellation Rates**

```python
df['Status_clean'] = df['Status'].astype(str).str.lower()
df['Is_Cancelled'] = df['Status_clean'].apply(lambda x: 1 if 'cancel' in x else 0)
df['Is_Returned']   = df['Status_clean'].apply(lambda x: 1 if 'return' in x else 0)

# Aggregate by Category
returns_cancel_summary = df.groupby('Category').agg(
    Total_Orders=('Order ID','count'),
    Cancelled_Orders=('Is_Cancelled','sum'),
    Returned_Orders=('Is_Returned','sum')
).reset_index()

# Calculate return & cancel rates
returns_cancel_summary['Cancellation_Rate_%'] = (returns_cancel_summary['Cancelled_Orders'] / returns_cancel_summary['Total_Orders']) * 100
returns_cancel_summary['Return_Rate_%'] = (returns_cancel_summary['Returned_Orders'] / returns_cancel_summary['Total_Orders']) * 100

returns_cancel_summary = returns_cancel_summary.sort_values(by='Cancellation_Rate_%', ascending=False)

print("ðŸ“¦ Customer Satisfaction Analysis â€” Cancellation & Returns")
returns_cancel_summary
```

**15. Identifying problematic products/categories to enhance quality**

```python
df['Status_clean'] = df['Status'].astype(str).str.lower()
df['Is_Cancelled'] = df['Status_clean'].apply(lambda x: 1 if 'cancel' in x else 0)
df['Is_Returned']  = df['Status_clean'].apply(lambda x: 1 if 'return' in x else 0)

problem_products = df.groupby(['Category', 'SKU']).agg(
    Total_Orders=('Order ID','count'),
    Cancelled=('Is_Cancelled','sum'),
    Returned=('Is_Returned','sum'),
    Total_Units_Sold=('Qty','sum')
).reset_index()

# Calculate rates
problem_products['Cancellation_Rate_%'] = (problem_products['Cancelled']/problem_products['Total_Orders'])*100
problem_products['Return_Rate_%'] = (problem_products['Returned']/problem_products['Total_Orders'])*100

problem_products = problem_products.sort_values(by=['Cancellation_Rate_%','Return_Rate_%'], ascending=False)

print("ðŸš¨ Problematic Products & Categories (High Cancellation/Return Rates)")
problem_products.head(20)
```
